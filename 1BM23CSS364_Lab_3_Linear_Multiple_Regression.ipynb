Linear Regression on Housing Area Price (given python code)

```python
"""Linear-Regression-Housing_Area_Price.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1CAlZml-P6V2V1RIrodgMfF8L3Ux4V9FT
"""

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('housing_area_price.csv')
df
```

```python
# Commented out IPython magic to ensure Python compatibility.
# matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)

"""(1) Predict price of a home with area = 3300 sqr ft"""

display(reg.predict(pd.DataFrame([[3300]], columns=['area'])))

"""reg.coef_

reg.intercept_"""

"""Y = m * X + b (m is coefficient and b is intercept)"""


"""(1) Predict price of a home with area = 5000 sqr ft"""
display(reg.predict(pd.DataFrame([[5000]], columns=['area'])))
```

Multiple Linear Regression (given python code)

```python
"""Multiple_LR_HomePrice.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1fK78C8TPV44HdvT6lsMhaau2wMtKXquQ
"""

import pandas as pd
import numpy as np
from sklearn import linear_model

df = pd.read_csv('homeprices_Multiple_LR.csv')
df

```

```python
"""Data Preprocessing: Fill NA values with median value of a column"""

df.bedrooms.median()

df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df

reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)

reg.coef_

reg.intercept_

"""Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old"""

reg.predict(pd.DataFrame([[3000, 3, 40]], columns=['area', 'bedrooms', 'age']))

112.06244194*3000 + 23388.88007794*3 + -3231.71790863*40 + 221323.00186540384
```

Predict the salary of an employee that has 12 years of experience

```python
df=pd.read_csv("salary.csv")
df.head()
df.info()
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df=df.dropna(subset=["YearsExperience"])
df
```

```python
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df['YearsExperience'],df['Salary'],color='red',marker='+')
```

```python
new_df=df.drop('Salary', axis='columns')

salary=df['Salary']

reg = linear_model.LinearRegression()
reg.fit(new_df, salary)

display(reg.predict(pd.DataFrame([[12]], columns=['YearsExperience'])))
```

Multi-variate/ Mulitple Linear Regression - Hiring

```python
df=pd.read_csv("hiring.csv")
df.head()
df.info()
```

```python
missing_values=df.isnull().sum()
missing_values
```

```python
df=df.dropna(subset=['experience', 'test_score(out of 10)'])
```

```python
from word2number import w2n

# Convert 'experience' column from words to numbers
df.experience = df.experience.apply(w2n.word_to_num)
df
```
